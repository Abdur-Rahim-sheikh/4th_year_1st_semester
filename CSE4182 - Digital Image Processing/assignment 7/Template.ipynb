```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
ROW = 2
COL = 4
```

```python
def img_plot(img,title,ind,cmap='gray'):
    plt.subplot(ROW,COL,ind*2 + 1)
    plt.title(title)
    plt.imshow(img,cmap=cmap)
    
    plt.subplot(ROW,COL,ind*2+2)
    plt.hist(img.ravel(),255,[0,255])
    plt.savefig('problem_7.jpg')
```

```python
img = plt.imread('assignment 7/vulture.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) + 35
```

```python
range_img = np.copy(gray)
range_img[gray<50] = 50
range_img[gray>200] = 200
imgs = [gray,gray+30,gray-30,range_img]
titles = ['original','right_shifted','left_shifted','range_value']
cnt = 0

plt.figure(figsize=(20,20))
for img,title in zip(imgs,titles):
    img_plot(img,title,cnt)
    cnt += 1
```
