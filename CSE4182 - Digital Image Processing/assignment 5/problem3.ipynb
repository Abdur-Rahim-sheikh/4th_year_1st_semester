```python
import cv2
import matplotlib.pyplot as plt
from numpy import random
import numpy as np
```

```python
dear = plt.imread('assignment 5/dear.jpeg')
print("dear dimentions: ",dear.shape)
plt.imshow(dear)
```

```python
def apply_kernel(img,kernel,ind,title):
    x,y = kernel.shape
    filtered = []
    for i in range(img.shape[0]-x+1):
        row = []
        for j in range(img.shape[1]-y+1):
            calc = np.sum(np.multiply(kernel,img[i:i+x,j:j+y]))
            if calc<0:
                calc = 0
            elif calc>255:
                calc = 255
            row.append(calc)

        filtered.append(row)

    filtered_hen = np.array(filtered)
    plt.subplot(1,2,ind)
    plt.title(title,fontsize=10)
    plt.imshow(filtered_hen,cmap='gray')
    plt.savefig('problem_3- kernel_apply.jpeg')
    
```

```python
gray_dear = cv2.cvtColor(dear,cv2.COLOR_RGB2GRAY)
sobel = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])
laplacian = np.array([[0, 1, 0],[1, -4, 1],[0, 1, 0]])
plt.figure(figsize=(20,22))
# apply_kernel(gray_dear,sobel,1,'sobel filter')
apply_kernel(gray_dear,laplacian,2,'laplacian filter')
```
