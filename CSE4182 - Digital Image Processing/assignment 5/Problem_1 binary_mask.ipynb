```python
import cv2
import matplotlib.pyplot as plt
from numpy import random
import numpy as np
```

```python
def show_image(img,title,ind,cmap=None):
    
    plt.subplot(1,3,ind)
    plt.title(title,fontsize=20)
    plt.imshow(img,cmap=cmap)
    plt.savefig('problem_1_binary_masking.jpeg')
```

```python
dear = plt.imread('assignment 5/dear.jpeg')
print(dear.shape[:2])
mask = np.zeros(dear.shape[:2],dtype="uint8")
mask[50:220,100:330] = 255

masked_dear = cv2.bitwise_and(dear,dear,mask=mask)
```

```python
masked2 = np.dstack((mask,mask,mask))
masked_dear2 = dear & masked2
```

```python
# dear = plt.imread('dear.jpeg')
print("dear dimentions: ",dear.shape)
plt.figure(figsize=(20,20))
show_image(dear,'original',1)

show_image(mask,'mask',2,cmap='gray')

show_image(masked_dear2,'after masking',3)
```
