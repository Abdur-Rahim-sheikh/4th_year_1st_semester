```python
def bit_slice(img,bit):
    img = img & (1<<bit)
    img = img > 0
    return img
```

```python
def plot_img(img,ind,title,cmap='binary'):
    plt.subplot(3,3,ind)
    plt.title(title,fontsize=15)
#     multiply them by 255 to make it high and low
    plt.imshow(img*255,cmap=cmap)
    plt.savefig('problem_2.jpeg')
    
```
